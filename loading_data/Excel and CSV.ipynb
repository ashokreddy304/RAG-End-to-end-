```python
import os
import pandas as pd
```

- Option 1: Converting table rows to text: This approach transforms each table row into a natural language description. Column headers become context labels, and values are integrated into readable text
- Option 2: Embedding tables in prompts:
As an alternative to converting rows into individual text snippets, you can embed the entire table directly into the LLM prompt. This works best for small tables whose full context fits into the model’s context window.

```python
file_path = "../datasets/csv_files/census-income.xlsx"
df_excel = pd.read_excel(io=file_path)
```

```python
df_excel.head()
```

```python
def create_text_description_of_row(row):
    row["text_description"] = (
        f"""The candidate {row['age']} years old is working in the
            {row['workclass']} sector. The candidate was born in
            {row['native-country']}, is {row['marital-status']}
            and has a {row['relationship']} relationship.
            The candidate has a {row['education']} degree and is
            working as a {row['occupation']}. The income of the
            candidate is {row['income']}."""
    )

    return row

df_extended = df_excel.apply(create_text_description_of_row, axis=1)
```
